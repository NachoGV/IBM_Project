## Feature Engineering

As there are no missing values, we will not need to impute any data.

The following changes will be made:
- Ensure types are either integer, float or category
- One hot encoding of color column for its use in regression - Keeping the original column for its use in classification
- Normalization of all features - No the taget columns

```python
# Load the dataset
wine_quality = fetch_ucirepo(id=186) 
wines = pd.DataFrame(wine_quality.data.original)
wines
```

```python
# One hot encoding
wines = pd.get_dummies(wines, columns=['color'])
wines['color'] = wine_quality.data.original['color']
wines
```

```python
# Normalize data
scaler = StandardScaler()
wines.iloc[:, :-4] = scaler.fit_transform(wines.iloc[:, :-4])
wines
```

```python
# Types
wines['color'] = wines['color'].astype('category')
wines.dtypes
```

```python
# Store the dataset
wines.to_csv('./data/wine_quality_preprocessed.csv', index=False)
```
